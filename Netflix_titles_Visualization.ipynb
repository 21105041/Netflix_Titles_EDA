```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Netflix_titles.csv")
print(data)
```

```python
data.info()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set plot style
sns.set(style="whitegrid")

# Content Distribution: Movies vs TV Shows
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=data, palette='Set1')
plt.title('Content Distribution: Movies vs TV Shows', fontsize=16)
plt.xlabel('Type of Content')
plt.ylabel('Count')
plt.show()
```

```python
#  Rating Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=data, palette='cool', order=data['rating'].value_counts().index)
plt.title('Distribution of Ratings on Netflix', fontsize=16)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Genre-wise Content Count
plt.figure(figsize=(10, 6))
genre_counts = data['listed_in'].str.split(',').explode().str.strip().value_counts().head(10)
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='Set1')
plt.title('Top 10 Most Common Genres on Netflix', fontsize=16)
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
# Country-wise Content Distribution (Top 10 Countries)
plt.figure(figsize=(12,8))
country_counts = data['country'].value_counts().head(10)
sns.barplot(x=country_counts.values, y=country_counts.index, palette='magma')
plt.title('Top 10 Countries with the Most Netflix Titles', fontsize=16)
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```python
# Split multiple directors (if any) and count occurrences
directors = data['director'].str.split(',').explode().str.strip()

# Get the top 10 most popular directors
director_counts = directors.value_counts().head(10)

# Plotting the most popular directors
plt.figure(figsize=(12, 6))
sns.barplot(x=director_counts.values, y=director_counts.index, palette='cool')
plt.title('Top 10 Most Popular Directors on Netflix', fontsize=16)
plt.xlabel('Number of Titles Directed')
plt.ylabel('Director')
plt.show()
```

```python
#  Distribution of Netflix Titles by Release Year
plt.figure(figsize=(10, 6))
sns.countplot(x='release_year', data=data, palette='viridis', order=data['release_year'].value_counts().index)
plt.title('Number of Titles Added to Netflix by Year', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
